```python
import pandas as pd
import pylab as pl
import numpy as np
import seaborn as sns
%matplotlib inline

import yaml
from nnaps import predictors
from sklearn import metrics
```

```python
#df = pd.read_csv('../data/sdBShortP_training_set.csv')
df = pd.read_csv('../data/sdBShortP_training_set_ce_applied.csv')
```

# Predicting He ignition

```python
setup = """
features: 
    - M1_init
    - FeH_init
    - M1_MLend
    - M1core_MLend
regressors: []
classifiers: 
    HeCoreBurning:
        loss: binary_crossentropy
model:
   - {'layer':'Dense',   'args':[200], 'kwargs': {'activation':'relu', 'name':'FC_1'} }
   - {'layer':'Dense',   'args':[100], 'kwargs': {'activation':'relu', 'name':'FC_2'} }
   - {'layer':'Dense',   'args':[50],  'kwargs': {'activation':'relu', 'name':'FC_3'} }
   
optimizer: adam
"""
setup = yaml.safe_load(setup)
```

```python
predictor_ignition = predictors.FCPredictor(setup=setup, data=df_stable)
```

```python
predictor_ignition.fit(epochs=100, batch_size=1024, reduce_lr=True)
```

```python
pl.figure(figsize=(10,6))
predictor_ignition.print_score()
predictor_ignition.plot_training_history()
```

```python
pl.figure(figsize=(8, 8))
predictor_ignition.plot_confusion_matrix()
```

```python
predictor_ignition.predict(df_ce)['HeCoreBurning'].values
```

```python
df_ce.loc[:,'HeCoreBurning'] = predictor_ignition.predict(df_ce)['HeCoreBurning'].values
```
