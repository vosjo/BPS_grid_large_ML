# Using all systems with a mass up to 2.5 Msol

```python
df = pd.read_csv('../data/raw/sdBShortP_large_BPS_set.csv')
```

```python
df.info(verbose=True)
```

```python
# extract the alpha value from the ce_parameters (alpha_ce is always the same as alpha_th)
df['alpha_ce'] = df['ce_parameters'].apply(lambda x: eval(x)['a_ce'])
```

```python
# Mark systems with a He ignition or burning error as error. ignore other errors.
df['error'] = df['error_flags'].apply(lambda x: 1 if 4 in eval(x) or 5 in eval(x) else 0)
```

```python
# remove systems that merge or are contact systems (the latter are likely also mergers)
df = df[(df['stability'] != 'merger') & (df['stability'] != 'contact')]
```

```python
df['product'] = 'UK'

for i, line in df.iterrows():
    
    sdA = line['sdA']
    sdB = line['sdB']
    sdO = line['sdO']
    HeBurn = line['HeCoreBurning']
    HeWD = line['He-WD']
    
    product = 'failed'
    if line['stability'] == 'CE' or line['stability'] == 'merger' or line['stability'] == 'contact':
        product = 'UK'
    elif sdA:
        product = 'sdA'
    elif sdB:
        product = 'sdB'
    elif sdO:
        product = 'sdO'
    elif HeWD:
        product = 'He-WD'
    elif HeBurn:
        product = 'HB'
    
    df.loc[i, 'product'] = product
```

```python
df['stability'].value_counts()
```

## create dataframes for the CE and stable systems, and the stable systems that ignite He

```python
df_ce =df[df['stability'] == 'CE']
df_stable = df[(df['stability'] == 'stable') & (df['error'] == 0)]
df_he = df_stable[df_stable['HeCoreBurning'] == 1]
```

# Predicting He ignition

The first method works well in predicting He ignition in stable systems. However, we don't know for sure if the parameters taken from the moment of maximum mass loss make sense for the CE systems for which we don't have a full detailed description of the mass loss phase. 

The alternative method uses the core and envelope mass after the mass loss ends, which we can get for both the stable and CE systems. The accuracy of this methods is similar to that of the first method.

```python
features = ['M1_init', 'FeH_init', 'M1_MLend', 'M1core_MLend']
df_select = df_stable[~df_stable[features].isna().any(axis=1)]
```

```python
setup = """
features: 
    - M1_init
    - FeH_init
    - M1_MLend
    - M1core_MLend
regressors: []
classifiers: 
    HeCoreBurning:
        processor: OrdinalEncoder
        loss: binary_crossentropy
model:
   - {'layer':'Dense',   'args':[200], 'kwargs': {'activation':'relu', 'name':'FC_1'} }
   - {'layer':'Dense',   'args':[100], 'kwargs': {'activation':'relu', 'name':'FC_2'} }
   - {'layer':'Dense',   'args':[50],  'kwargs': {'activation':'relu', 'name':'FC_3'} }
   
optimizer: adam
"""
setup = yaml.safe_load(setup)
```

```python
predictor_ignition = predictors.FCPredictor(setup=setup, data=df_select)
```

```python
predictor_ignition.fit(epochs=100, batch_size=256, reduce_lr=True)
```

```python
pl.figure(figsize=(10,6))
predictor_ignition.print_score()
predictor_ignition.plot_training_history()
```

```python
pl.figure(figsize=(8, 8))
predictor_ignition.plot_confusion_matrix()
```

```python
predictor_ignition.save_model('model_he_ignition.h5')
```

# Predicting core and envelope mass during He core burning

In the stable systems, the core continues growing between the moment of mass loss end, and the He iginition. We want to model this change so that we can more accurately predict the He core burning properties in the post-CE systems

Experiments were made with predicting both Mcore and Menv, but it turns out that predicting only the core and then getting the envelope mass by subtracting the core from the total mass is much more efficient and accurate.

First we remove some systems from the training run for which the mass loss phase is not accurately summarized. These systems have a second mass loss phase before He ignition. 

```python
sns.relplot('M1core_MLend', 'M1core_HeCoreBurning', data=df_he, hue='M1_init', size='M1_init')
pl.plot([0.25, 0.5], [0.25, 0.5], '--k')
```

```python
df_he['core_growth'] = df_he['M1core_HeCoreBurning'] - df_he['M1core_MLend']
```

```python
features = ['M1_init', 'FeH_init', 'M1_MLend', 'M1core_MLend']
df_select = df_he[~df_he[features].isna().any(axis=1)]

# remove systems that lose core mass, as those have their mass loss phase after the He core buring phase.
df_select['core_growth'] = df_select['M1core_HeCoreBurning'] - df_select['M1core_MLend']
df_select = df_select[df_select['core_growth'] > 0]
```

```python
setup = """
features: 
    - M1_init
    - FeH_init
    - M1_MLend
    - M1core_MLend
regressors:
    M1core_HeCoreBurning:
        processor: StandardScaler
classifiers: []
model:
   - {'layer':'Dense',   'args':[300], 'kwargs': {'activation':'relu', 'name':'FC_1'} }
   - {'layer':'Dense',   'args':[200],  'kwargs': {'activation':'relu', 'name':'FC_2'} }
   - {'layer':'Dense',   'args':[100],  'kwargs': {'activation':'relu', 'name':'FC_3'} }
   - {'layer':'Dense',   'args':[100],  'kwargs': {'activation':'relu', 'name':'FC_4'} }
   
optimizer: adam
"""
setup = yaml.safe_load(setup)
```

```python
predictor_coremass = predictors.FCPredictor(setup=setup, data=df_select)
```

```python
predictor_coremass.fit(epochs=100, batch_size=256, reduce_lr=True)
```

```python
pl.figure(figsize=(10,6))
predictor_coremass.print_score()
predictor_coremass.plot_training_history()
```

```python
pred = predictor_coremass.predict(df_select)
df_select.loc[:,'M1core_HeCoreBurning_pred'] = pred['M1core_HeCoreBurning'].values
df_select.loc[:,'M1core_HeCoreBurning_err'] = np.abs(df_select['M1core_HeCoreBurning'] - df_select['M1core_HeCoreBurning_pred'])
```

```python
pl.figure(figsize=(12,10))

pl.subplot(221)
sns.scatterplot('M1core_HeCoreBurning', 'M1core_HeCoreBurning_pred', data = df_select)
pl.plot([0.3, 0.5], [0.3, 0.5], '--k')

pl.subplot(222)
sns.scatterplot('M1core_MLend', 'M1core_HeCoreBurning', data=df_select, hue='M1core_HeCoreBurning_err')
pl.plot([0.25, 0.5], [0.25, 0.5], '--k')

pl.subplot(223)
sns.scatterplot('M1core_HeCoreBurning', 'M1core_HeCoreBurning_err', data=df_select)

pl.subplot(224)
sns.scatterplot('core_growth', 'M1core_HeCoreBurning_err', data=df_select)
pl.plot([0., 0.1], [0., 0.1], '--k')
```

```python
predictor_coremass.save_model('model_coremass_evolution.h5')
```

# Predicting subdwarf vs HB

It turns out that to estimate atmospheric properties during the core He burning phase, it is necessary to differentiate between subdwarfs and horizontal branch objects.

```python
features = ['M1_init', 'FeH_init', 'M1core_HeCoreBurning', 'M1env_HeCoreBurning']
df_select = df_he[~df_he[features].isna().any(axis=1)]

# remove systems that lose core mass, as those have their mass loss phase after the He core buring phase.
df_select['is_sd'] = df_select['product'].apply(lambda x: 0 if x == 'HB' else 1)
```

```python
setup = """
features: 
    - M1_init
    - FeH_init
    - M1core_HeCoreBurning
    - M1env_HeCoreBurning
regressors: []
classifiers: 
    is_sd:
        loss: binary_crossentropy
model:
   - {'layer':'Dense',   'args':[100], 'kwargs': {'activation':'relu', 'name':'FC_1'} }
   - {'layer':'Dense',   'args':[75],  'kwargs': {'activation':'relu', 'name':'FC_2'} }
   - {'layer':'Dense',   'args':[50],  'kwargs': {'activation':'relu', 'name':'FC_3'} }
   
optimizer: adam
"""
setup = yaml.safe_load(setup)
```

```python
predictor2 = predictors.FCPredictor(setup=setup, data=df_select)
```

```python
predictor2.fit(epochs=100, batch_size=256, reduce_lr=True)
```

```python
pl.figure(figsize=(10,6))
predictor2.print_score()
predictor2.plot_training_history()
```

```python
pl.figure(figsize=(8, 8))
predictor2.plot_confusion_matrix()
```

```python
predictor2.save_model('model_sd_vs_hb.h5')
```

# Predicting He core burning properties

```python
features = ['M1_init', 'FeH_init', 'M1core_HeCoreBurning', 'M1env_HeCoreBurning']
df_select = df_he[~df_he[features].isna().any(axis=1)]

# add a flag to indicate subdwarf
df_select['is_sd'] = df_select['product'].apply(lambda x: 0 if x == 'HB' else 1)

# only keep HB stars with a radius less than 4
df_select = df_select[df_select['rad_HeCoreBurning'] < 4.0]

# oversample sds
df_train = pd.concat([df_select[df_select['is_sd'] == 1].sample(1000, replace=True),
                      df_select[df_select['is_sd'] == 0].sample(500, replace=True)])
df_train = df_train.sample(len(df_train))
```

```python
predictor_heproperties.fit(epochs=100, batch_size=256, reduce_lr=True)
```

```python
pl.figure(figsize=(15,10))
predictor_heproperties.print_score()
predictor_heproperties.plot_training_history()
```

```python
pred = predictor_heproperties.predict(df_select)

try:
    df_select['teff_predicted'] = pred['teff_HeCoreBurning'].values
    df_select['teff_diff'] =  np.abs(df_select['teff_HeCoreBurning'] - df_select['teff_predicted']) / df_select['teff_HeCoreBurning']

    print('Teff:')
    print(df_select.groupby('product')['teff_diff'].mean())
except Exception:
    pass

try:
    df_select['logg_predicted'] = pred['logg_HeCoreBurning'].values
    df_select['logg_diff'] = np.abs(df_select['logg_HeCoreBurning'] - df_select['logg_predicted']) / df_select['logg_HeCoreBurning']
    
    print('logg:')
    print(df_select.groupby('product')['logg_diff'].mean())
except Exception:
    pass    

try:
    df_select['rad_predicted'] = pred['rad_HeCoreBurning'].values
    df_select['rad_diff'] = np.abs(df_select['rad_HeCoreBurning'] - df_select['rad_predicted']) / df_select['rad_HeCoreBurning']

    print('rad:')
    print(df_select.groupby('product')['rad_diff'].mean())
except Exception:
    pass    
```

# Predicting He core burning properties separately for HB and SDs

It is way more efficient to predict the atmospheric properties of SDs and HBs with different

```python
data = df_he.copy()
data['is_sd'] = predictor2.predict(data)['is_sd'].values
data['is_hb'] = data['is_sd'].apply(lambda x: 0 if x == 1 else 1)
data_hb = data[data['is_hb'] == 1].reset_index(drop=True)
data_sd = data[data['is_sd'] == 1].reset_index(drop=True)
```

```python
setup = """
features: 
    M1_init:
    FeH_init:
    M1core_HeCoreBurning:
    M1env_HeCoreBurning:
regressors:
    teff_HeCoreBurning:
        processor: StandardScaler
    logg_HeCoreBurning:
        processor: StandardScaler
    rad_HeCoreBurning:
        processor: StandardScaler
classifiers: []
model:
   - {'layer':'Dense',   'args':[500], 'kwargs': {'activation':'relu', 'name':'FC_1'} }
   - {'layer':'Dropout', 'args':[0.3], 'kwargs': {'name':'DO_1'} }
   - {'layer':'Dense',   'args':[300], 'kwargs': {'activation':'relu', 'name':'FC_2'} }
   - {'layer':'Dropout', 'args':[0.3], 'kwargs': {'name':'DO_2'} }
   - {'layer':'Dense',   'args':[100], 'kwargs': {'activation':'relu', 'name':'FC_3'} }
optimizer: adam
"""
setup = yaml.safe_load(setup)
```

```python
predictor_atm_hb = predictors.FCPredictor(setup=setup, data=data_hb)
predictor_atm_sd = predictors.FCPredictor(setup=setup, data=data_sd)
```

```python
predictor_atm_hb.fit(epochs=100, batch_size=1024, reduce_lr=True)
```

```python
predictor_atm_sd.fit(epochs=100, batch_size=1024, reduce_lr=True)
```

```python
pl.figure(figsize=(15,10))
print('HB')
predictor_atm_hb.print_score()
predictor_atm_hb.plot_training_history()
pl.figure(figsize=(15,10))
print('SD')
predictor_atm_sd.print_score()
predictor_atm_sd.plot_training_history()
```

```python
predictor_atm_hb.save_model('model_hb_properties.h5')
predictor_atm_sd.save_model('model_sd_properties.h5')
```
