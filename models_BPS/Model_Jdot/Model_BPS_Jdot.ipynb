```
import pandas as pd
import pylab as pl
import numpy as np

import yaml
```

```
from nnaps.predictors import FCPredictor
```

```
!ls /datasets/sdbshortp_large_bps_data
```

```
datadir = '/datasets/sdbshortp_large_bps_data/'
d1 = pd.read_csv(datadir+'sdBShortP_large_BPS_set_stability_jdotp_VAR_alphace_VAR_ce_applied.csv')
d2 = pd.read_csv(datadir+'sdBShortP_large_BPS_set_stability_jdotp_VAR_alphace_VAR_ce_focus_ce_applied.csv')
data = pd.concat([d1, d2])
```

```
data.info(verbose=True)
```

```
data['alpha_ce'] = data['ce_parameters'].apply(lambda x: eval(x)['a_ce'])
data['alpha_th'] = data['ce_parameters'].apply(lambda x: eval(x)['a_th'])
```

```
data['product'] = data['product'].astype(str)
```

```
setup = """
features: 
    - M1_init
    - q_init
    - P_init
    - FeH_init
    - stability_limit
    - alpha_ce
    - alpha_th
regressors:
    M1_final:
        processor: StandardScaler
    q_final:
        processor: StandardScaler
    P_final:
        processor: StandardScaler
classifiers: 
    product:
        
model:
   - {'layer':'Dense',   'args':[500], 'kwargs': {'activation':'relu', 'name': 'FC1'} }
   - {'layer':'Dense',   'args':[300], 'kwargs': {'activation':'relu', 'name': 'FC2'} }
   - {'layer':'Dense',   'args':[200], 'kwargs': {'activation':'relu', 'name': 'FC3'} }
   - {'layer':'Dense',   'args':[100], 'kwargs': {'activation':'relu', 'name': 'FC4'} }
   - {'layer':'Dense',   'args':[50],  'kwargs': {'activation':'relu', 'name': 'FC5'} }
optimizer: adam
"""
setup = yaml.safe_load(setup)
predictor = FCPredictor(setup=setup, data=data)
```

```
predictor.fit(epochs=25, batch_size=2048, reduce_lr=True)
```
